```
import scipy.integrate
import numpy as np
import matplotlib.pyplot as plt
```

```
def SIR_model(y, t, beta, gamma):
  S, I, R = y
  dsdt = -beta*I*S/N0 
  didt = beta*I*S/N0 - gamma*I
  drdt = gamma*I
  return ([dsdt, didt, drdt])
```

```
N= 300000
N0, S0, I0, R0, beta, gamma = N, N-10, 10, 0.0,  0.2, 1/14

t = np.linspace(0, 365, 365*24)

solution = scipy.integrate.odeint(SIR_model, [S0, I0, R0], t, args=(beta, gamma))
solution = np.array(solution)
```

```
plt.figure(figsize=[12,8])
plt.plot(t, solution[:,0], label="S(t)", color="blue", linestyle="-")
plt.plot(t, solution[:,1], label="I(t)", color="red", linestyle="-")
plt.plot(t, solution[:,2], label="R(t)", color="green", linestyle="-")
plt.plot(t, solution[:,0]+solution[:,1]+solution[:,2], label="N(t)", color="gray", linestyle="-")
plt.grid()
plt.legend()
plt.ylabel("Proporciones")
plt.xlabel("Tiempo")
plt.title("Modelo SIR")
plt.show()
```

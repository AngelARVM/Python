```
import numpy as np
import matplotlib as plt
```

```
lamb, mu, beta, alpha, gamma = 0, 0, 0.025, 1/(7*24), 1/(14*24) 

N,S,E,I,R,r0 = 44555666, 44555615, 1, 0, 0,0
```

```
periodo = np.array(list(range(0,180)))
datos_S = np.zeros(180)
datos_E = np.zeros(180)
datos_I = np.zeros(180)
datos_R = np.zeros(180)
```

```
def dsdt():
  global S
  S+= lamb-mu*S-(beta*S*(I/N))

def dedt():
  global E
  E+= (beta*S*(I/N))-(mu+alpha)*E

def didt():
  global I
  I+= alpha*E-(gamma+mu)*I

def drdt():
  global R
  R+= gamma*I-mu*R
```

```
for i in periodo:
  for j in range (0,24):
    dsdt()
    dedt()
    didt()
    drdt()
  r0=(alpha/(mu+alpha))*(beta/(mu+gamma))   
  
  datos_S[i] = round(S)
  datos_E[i] = round(E)
  datos_I[i] = round(I)
  datos_R[i] = round(R)
  print (round(datos_S[i]),round(datos_E[i]),round(datos_I[i]),round(datos_R[i]),round(datos_S[i])+round(datos_E[i])+round(datos_I[i])+round(datos_R[i]))
```

```
plt.pyplot.figure(figsize=(10, 8), dpi=80)
plt.pyplot.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.pyplot.plot(periodo,datos_S, color="green", linewidth=2.5, linestyle="-", label="Suceptible")
plt.pyplot.plot(periodo,datos_E, color="orange", linewidth=2.5, linestyle="--", label="Expuesto")
plt.pyplot.plot(periodo,datos_I, color="red", linewidth=2.5, linestyle="-.", label="Infectado")
plt.pyplot.plot(periodo,datos_R, color="blue", linewidth=2.5, linestyle=":", label="Recuperado")
plt.pyplot.show()
```
